```python
df = df[df['State']=='OH']
df1 = df[['Severity','Zipcode','Weather_Condition']]
df1 = df1.dropna()
df1['Zipcode'] = df1['Zipcode'].str.slice(stop=5)
df1 = df1.dropna()
df1['Weather_Condition'] = df1['Weather_Condition'].map({'Light Rain': 1, 'Overcast': 2, 'Fair': 0, 'Thunder': 3, 'Partly Cloudy': 4,'Mostly Cloudy': 5, 'Snow': 6, 'Light Snow': 7, 'Snow': 8, 'Clear': 9})
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(df1)
kmeans.labels_
```
