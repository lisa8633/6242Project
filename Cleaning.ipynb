```python
import sqlite3
from sqlite3 import Error
import csv
import pandas as pd
SHOW = True
```

```python
df = pd.read_csv('US_Accidents_Dec20_updated.csv')
df
```

```python
df1 = df[['ID','Start_Time','Severity','Start_Lat','Start_Lng','Zipcode','Weather_Condition','Temperature(F)','Humidity(%)','Visibility(mi)','Amenity','Bump','Crossing','Give_Way','Junction','No_Exit','Railway','Roundabout','Station','Stop','Traffic_Calming','Traffic_Signal','Turning_Loop','Sunrise_Sunset']]
df1['Zipcode'] = df1['Zipcode'].str.slice(stop=5).astype(float)
df1["Amenity"] = df1["Amenity"].astype(int)
df1["Bump"] = df1["Bump"].astype(int)
df1["Crossing"] = df1["Crossing"].astype(int)
df1["Give_Way"] = df1["Give_Way"].astype(int)
df1["Junction"] = df1["Junction"].astype(int)
df1["No_Exit"] = df1["No_Exit"].astype(int)
df1["Railway"] = df1["Railway"].astype(int)
df1["Roundabout"] = df1["Roundabout"].astype(int)
df1["Station"] = df1["Station"].astype(int)
df1["Stop"] = df1["Stop"].astype(int)
df1["Traffic_Calming"] = df1["Traffic_Calming"].astype(int)
df1["Traffic_Signal"] = df1["Traffic_Signal"].astype(int)
df1["Turning_Loop"] = df1["Turning_Loop"].astype(int)
df1['Start_Time']= pd.to_datetime(df['Start_Time'])
a = {'Night' : 1,'Day' : 0}
df1['Sunrise_Sunset'] = df1['Sunrise_Sunset'].map(a)
df1
```

```python
dataTypeSeries = df1.dtypes
print('Data type of each column of Dataframe :')
print(dataTypeSeries)
```

```python
weather = pd.DataFrame(df1['Weather_Condition'], columns = ['Weather_Condition'])
one_hot = pd.get_dummies(weather['Weather_Condition'])
one_hot
```

```python
df1= df1.drop('Weather_Condition', axis=1)
dataTypeSeries = df1.dtypes
print('Data type of each column of Dataframe :')
print(dataTypeSeries)
```

```python
df1 = df1.join(one_hot)
dataTypeSeries = df1.dtypes
print('Data type of each column of Dataframe :')
print(dataTypeSeries)
```
