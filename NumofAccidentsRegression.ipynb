```python
#Convert Data Type
df1=df_org[['ID','State','Temperature(F)','Visibility(mi)','Wind_Speed(mph)','Weather_Condition']]
rename_dict={'State':'state',
             'Temperature(F)':'temperature','Wind_Speed(mph)':'wind_speed',
             'Visibility(mi)':'visibility','Weather_Condition':'weather_condition'}
df1.rename(columns=rename_dict,inplace=True)
df1['date']=pd.to_datetime(df_org['Start_Time']).dt.date
df1.head(3)
```

```python
df1[df1.isna().any(axis=1)].shape[0]
df1[df1.weather_condition.isna()].shape[0]/df1.shape[0]
```

```python
df2=df1[~df1.weather_condition.isna()].copy()
df2.shape
```

```python
df2['rain'],df2['snow_ice'],df2['thunder'],df2['dust'],df2['fog'],df2['cloud'],df2['squalls'],df2['fair']=0,0,0,0,0,0,0,0
df2['extreme_weather']=0
df2['flag']=0
df2.head(3)
```

```python
weather_dict={'rain':'rain','showers in':'rain','drizzle':'rain',
              'snow':'snow_ice','wintry':'snow_ice','sleet':'snow_ice','ice':'snow_ice','hail':'snow_ice',
              'thunder':'thunder','t-storm':'thunder',
              'dust':'dust','sand':'dust','ash':'dust',
              'fog':'fog','mist':'fog','haze':'fog','smoke':'fog',
              'cloud':'cloud','overcast':'cloud',
              'squalls':'squalls','tornado':'squalls',
              'clear':'fair','fair':'fair','precipi':'fair'
             }
for index, row in df2.iterrows():
    if index%1000==0:
        print('Complete %d rows'%index)
    w=row['weather_condition'].lower()
    for k,v in weather_dict.items():
        if k in w:
            df2.loc[index,v]=1
            df2.loc[index,'flag']=1
    if 'heavy' in w or 'tornado' in w or 'squalls' in w:
        df2.loc[index,'extreme_weather']=1
df2.head()
```

```python
df2.reset_index(inplace=True,drop=True)
```

```python
df2[df2['flag']==0].weather_condition.unique()
```

```python
for index,row in df2[df2['flag']==0].iterrows():
    w=row['weather_condition'].lower()
    if 'cloud' in w:
        df2.loc[index,'cloud']=1
        df2.loc[index,'flag']=1
    if 'precipi' in w:
        df2.loc[index,'fair']=1
        df2.loc[index,'flag']=1
    if 'ash' in w:
        df2.loc[index,'dust']=1
        df2.loc[index,'flag']=1
```

```python
df2[df2['flag']==0]
```

```python
df2.to_csv("CleanData_df2.csv")  
```

```python
agg_dict={'ID':'count','temperature':['min','max'],'visibility':'min','wind_speed':'max',
          'rain':'max','snow_ice':'max','thunder':'max','dust':'max','fog':'max','cloud':'max',
          'squalls':'max','fair':'max','extreme_weather':'max'}
df3=df2.groupby('date').agg(agg_dict)
df3.head(3)
```

```python
df3.head(10)
```

```python
df3.shape
```

```python
df3.columns
```

```python
df3.columns = ['accident_counts','highest_temp','lowest_temp','min_visi','max_wind_speed',
               'is_rain','is_snow_ice','is_thunder','is_dust','is_fog','is_cloud','is_squalls','is_fair','is_extreme']
df3.head()
```

```python
from sklearn.model_selection import train_test_split
X=df3.iloc[:,1:]
y=df3.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=66)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))

```

```python
for x,y in zip(y_test,y_pred):
    print('%d'x,y)
```

<h3>Poisson Regression</h3>

```python
poisson_regr = linear_model.PoissonRegressor()
poisson_regr.fit(X_train,y_train)
y_pred_poisson=poisson_regr.predict(X_test)
poisson_score=poisson_regr.score(X_test,y_pred_poisson)
print(poisson_score)
```

```python
for a,b in zip(y_test,y_pred_poisson):
    print('%-10d%.2f'%(a,b))
```
